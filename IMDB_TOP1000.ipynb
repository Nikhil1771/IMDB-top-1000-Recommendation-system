```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
movie_title = []
movie_year = []
movie_length = []
movie_genre = []
movie_rating = []
movie_des = []
movie_crew = []
movie_poster = []
```

```python
number=['1','251','501','751']
for num in number:
    r = requests.get("https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=250&start="+num+"&ref_=adv_nxt")

    #convert into a beatuful soup object
    soup = bs(r.content)
    
    imdb_table = soup.find(class_ = "lister list detail sub-list")
    
    
    movie_title_coloumn = imdb_table.find_all(class_ = "lister-item-header")
    for row in movie_title_coloumn:
        title = row.a.text
        movie_title.append(title)
        
    
    movie_year_coloumn = imdb_table.find_all(class_ = "lister-item-year text-muted unbold")
    for row in movie_year_coloumn:
        year = row.text
        movie_year.append(year)
        
    
    movie_length_coloumn = imdb_table.find_all(class_ = "runtime")
    for row in movie_length_coloumn:
        length = row.text
        movie_length.append(length)
        
    
    movie_genre_coloumn = imdb_table.find_all(class_ = "genre")
    for row in movie_genre_coloumn:
        genre = row.text
        movie_genre.append(genre)
        
        
    movie_rating_coloumn = imdb_table.find_all(class_ = "inline-block ratings-imdb-rating")
    for row in movie_rating_coloumn:
        rating = row.strong.text
        movie_rating.append(rating)
        
        
    movie_des_coloumn = imdb_table.find_all('p',class_ = "text-muted" )
    for row in movie_des_coloumn[1::2]:
        des = row.text
        movie_des.append(des)
        
    
    movie_crew_column = imdb_table.find_all('p' , class_ = "")
    for row in movie_crew_column:
        crew = row.text
        movie_crew.append(crew)
        
    movie_poster_column = imdb_table.find_all("img") 
    for image in movie_poster_column:
        movie_poster.append(image['loadlate'])
```

```python
df['movie_year'] = df['movie_year'].apply(lambda x: x.replace('(' , '').replace(')','').replace('I' , '').replace('II' , '').replace('III' , ''))
```

```python
df['movie_genre'] = df['movie_genre'].apply(lambda x: x.replace('\n' , ''))
df['movie_des'] = df['movie_des'].apply(lambda x: x.replace('\n' , ''))
df['movie_length'] = df['movie_length'].apply(lambda x: x.replace('min' , ''))
df['movie_crew'] = df['movie_crew'].apply(lambda x: x.replace('\n' , ''))
```

```python
df[['director_name' , 'stars_name']] = df['movie_crew'].apply(lambda x : pd.Series(str(x).split('|')))
```

```python
df = df.drop(['movie_crew'],axis=1)
df['director_name'] = df['director_name'].apply(lambda x: x.replace('Director:' , '').replace(' ' , '').replace('Directors:' , ''))
df['stars_name'] = df['stars_name'].apply(lambda x: x.replace('Stars:' , '').replace(' ' , ''))
df
```

```python
from nltk.stem.porter import PorterStemmer
ps  = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# VECTORISATION

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english') 
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

# COSINE SIMILARITY

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
#findinding the index of the movie
new_df[new_df['movie_title'] == 'Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x : x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['movie_title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x : x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].movie_title)
```

```python
recommend('The Dark Knight')
```
